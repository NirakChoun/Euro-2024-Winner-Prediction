```python
import pandas as pd 

df_historical_data = pd.read_csv("euro_historical_data.csv")
df_fixture = pd.read_csv("euro_2024_fixtures.csv")

# Erase the unnecessary space in the Home and Away columns
df_fixture["Home"] = df_fixture["Home"].str.strip()
df_fixture["Away"] = df_fixture["Away"].str.strip()

# Delete (a.e.t) in the Score column
df_historical_data["Score"] = df_historical_data["Score"].str.replace(r"\(.*\)", "", regex=True).str.strip()
df_historical_data["Home"] = df_historical_data["Home"].str.strip()
df_historical_data["Away"] = df_historical_data["Away"].str.strip()

import pandas as pd 

# Read the CSV files
df_historical_data = pd.read_csv("euro_historical_data.csv")
df_fixture = pd.read_csv("euro_2024_fixtures.csv")

# Erase the unnecessary space in the Home and Away columns
df_fixture["Home"] = df_fixture["Home"].str.strip()
df_fixture["Away"] = df_fixture["Away"].str.strip()

# Clean up the Score, Home, and Away columns in the historical data
df_historical_data["Score"] = df_historical_data["Score"].str.replace(r"\(.*\)", "", regex=True).str.strip()
df_historical_data["Home"] = df_historical_data["Home"].str.strip()
df_historical_data["Away"] = df_historical_data["Away"].str.strip()

# Extract only the numeric scores from the Score column
score_pattern = r"(\d+)[–-](\d+)"  # Regex to match scores like "2-1" or "2–1"
scores = df_historical_data["Score"].str.extract(score_pattern)

# Assign the extracted scores to new columns and convert to integers
df_historical_data["HomeGoals"] = scores[0].astype(float).astype(int)
df_historical_data["AwayGoals"] = scores[1].astype(float).astype(int)

# Delete the score column
df_historical_data.drop("Score", axis=1, inplace=True)
df_historical_data.rename(columns={"Home" : "HomeTeam", "Away" : "AwayTeam"}, inplace=True)


# Create a total goals column
df_historical_data["TotalGoals"] = df_historical_data["HomeGoals"] + df_historical_data["AwayGoals"]


df_historical_data.to_csv("clean_euro_historical_data.csv", index=False)
df_fixture.to_csv("clean_euro_2024_fixtures.csv", index=False)

years = [1960, 1964, 1968, 1972, 1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020]

allMatches = 0

for year in years:
    allMatches += len(df_historical_data[df_historical_data["Year"]==year])

print(allMatches)

df_historical_data
```
